```python
kernmerkmale = pd.read_excel('demo_excel.xls', sheetname=2, header=3, index_col=1)
kernmerkmale.head()
```

```python
#Task4
bundeslaender = kernmerkmale[kernmerkmale['SATZART']==10]
bundeslaender
```

```python
#Task5
bundeslaender.plot(x=('NAME'),y=[('ALTER_1'),('ALTER_2'),('ALTER_3'),('ALTER_4'),('ALTER_5')],kind='barh')
```

```python
#Task6
relativ_age=bundeslaender[['ALTER_1','ALTER_2','ALTER_3','ALTER_4','ALTER_5']].div(bundeslaender['EWZ'],axis=0)
relativ_age.plot(x=bundeslaender['NAME'],kind='barh')
```

```python
#Task7
berlin = bundeslaender[bundeslaender['NAME'] == 'Berlin']
bundeslaender_rel= berlin [['ALTER_1','ALTER_2','ALTER_3','ALTER_4','ALTER_5']].div(bundeslaender['EWZ'],axis=0)
rel= bundeslaender_rel.dropna(axis=0,how ='all')

rel_inv_ber=rel.transpose()
rel_inv_ber.plot(subplots='TRUE', kind= 'pie')
```
